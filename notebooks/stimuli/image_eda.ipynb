```python
import numpy as np
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
import os
```

```python
imdir = '/media/timo/data/DPHIL_03_TRANSFERLEARN/datasets/src/AwA2-data/Animals_with_Attributes2/'
```

```python
os.listdir(imdir)
```

```python
# list of predicates
with open(imdir+'predicates.txt','r') as f:
    predicates = f.read()

# remove tabs and new lines
predicates = [p.split('\t')[-1] for p in predicates.split('\n')][:-1]

# list of animals 
with open(imdir+'classes.txt','r') as f:
    classes = f.read()
classes = [c.split('\t')[-1] for c in classes.split('\n')][:-1]
```

```python
print(predicates, '\n\n', classes)
print(len(predicates))
print(len(classes))
```

```python
# attributes (binary)
attributes = pd.read_csv(imdir+'predicate-matrix-binary.txt',sep=' ',names=predicates)
attributes.index = classes
```

```python
attributes = attributes[['small','big','slow','fast']]

# masks for four cases 
mask_small_slow = (attributes['small']==1) & (attributes['slow']==1)
print(sum(mask_small_slow))
mask_small_fast = (attributes['small']==1) & (attributes['slow']!=1)
print(sum(mask_small_fast))
mask_big_slow = (attributes['small']!=1) & (attributes['slow']==1)
print(sum(mask_big_slow))
mask_big_fast = (attributes['small']!=1) & (attributes['slow']!=1)
print(sum(mask_big_fast))
```

```python
# print names of each category
print('\nsmall and slow animals (1,1):\n',attributes.index[mask_small_slow].tolist())
# print('\nsmall and fast animals (1,2):\n',attributes.index[mask_small_fast].tolist())
# print('\nbig and slow animals (2,1):\n',attributes.index[mask_big_slow].tolist())
# print('\nbig and fast animals (2,2):\n',attributes.index[mask_big_fast].tolist())
```
